## Sentimen Analisis Review Aplikasi Tokopedia

### Import Library

```python
from google_play_scraper import app, reviews, Sort, reviews_all
import pandas as pd  # Pandas untuk manipulasi dan analisis data
```

```python
import nltk  # Import pustaka NLTK (Natural Language Toolkit).
nltk.download('punkt')  # Mengunduh dataset yang diperlukan untuk tokenisasi teks.
nltk.download('stopwords')  # Mengunduh dataset yang berisi daftar kata-kata berhenti (stop words) dalam berbagai bahasa.
```

### Scraping

```python
scrape_review= reviews_all(
  'com.tokopedia.tkpd',            # ID aplikasi
  lang='en',                  # defaults to ‚Äòen‚Äô
  country='sg',               # defaults to ‚Äòus‚Äô
  sort = Sort.MOST_RELEVANT,  # defaults to Sort.MOST_RELEVANT
  filter_score_with=None      # defaults to None (means all score)
)
```

```python
df = pd.DataFrame(scrape_review)
df
```

```python
df.info()
```

```python
# menghapus baris yang memiliki nilai yang hilang (NaN) dari df
clean_df = df.dropna()
clean_df.info()
```

```python
clean_df.to_csv("tokopedia_review.csv", index=None, header=True)
```
